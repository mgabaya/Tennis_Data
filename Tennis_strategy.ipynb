```python
serve = pd.read_csv('Mens/charting-m-stats-ServeBasics.csv', encoding='ISO-8859-1', index_col=False)
serve
```

```python
fed_serve = serve[serve['match_id'].str.contains('-Roger_Federer-')]
fed_serve
```

```python
fed_serve.info()
```

```python
fed_serve.reset_index(inplace=True, drop=True)
fed_serve.head(19)
```

```python
fed_serve.describe()
```

```python
#df['A'].replace('foo', 'qux', inplace=True)
aces = fed_serve[fed_serve['row'].str.contains('otal')]
aces['row'].replace('1 Total', 'Federer', inplace=True)
aces['row'].replace('2 Total', 'Competitor', inplace=True)
aces
```

```python
fig = px.box(
    aces,
    x='row',
    y='aces',
    title='Federer\'s Aces Vs the Competition' #,
    #width=width_in_pixels,
    #height=height_in_pixels
)

fig.show()
```
